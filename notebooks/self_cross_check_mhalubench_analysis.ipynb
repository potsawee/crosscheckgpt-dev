# SelfCheck

```python
selfcheck_scores_dict = {}
for llm in llms:
    with open(f"../outputs/mhalubench_batch1/{llm}.json") as f:
        x = json.load(f)
    items = []
    selfcheck_scores = []
    for item in x:
        if item['llm_gen_text'] != llm or item['llm_gen_evidence'] != llm:
            continue
        items.append(item)
        selfcheck_scores.append(np.mean(item['scores']))
    print(llm, np.mean(selfcheck_scores))
    selfcheck_scores_dict[llm] = np.mean(selfcheck_scores)
with open("../outputs/selfcheck_mhalubench.json", "w") as f:
    json.dump(selfcheck_scores_dict, f)
```

# CrossCheck

```python
crosscheck_scores_dict = {}
for llm in llms:
    with open(f"../outputs/mhalubench_batch1/{llm}.json") as f:
        x = json.load(f)
    items = []
    crosscheck_scores = []
    for item in x:
        items.append(item)
        crosscheck_scores.append(np.mean(item['scores']))
    print(llm, np.mean(crosscheck_scores))
    # print(llm, len(crosscheck_scores))
    crosscheck_scores_dict[llm] = np.mean(crosscheck_scores)
with open("../outputs/crosscheck_mhalubench.json", "w") as f:
    json.dump(crosscheck_scores_dict, f)
```
